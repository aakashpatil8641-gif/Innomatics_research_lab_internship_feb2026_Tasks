Problem 1: Unique Words in a Sentence

```python
sentence = input("Enter a Sentence :")

word = sentence.split()

unique_words = set(word)  #set remove duplicates


print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Highest Salary:", highest_employee, "-", highest_salary)
```

Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

total_days = len(attendance)

present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

Problem 7: Count Character Frequency

```python
text = "pythonp"

char_count = {}

for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```

Problem 9: Check if a Key Exists in a Dictionary

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

key_to_check = "Ravi"

if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

Problem 10: Calculate Average Marks

```python
marks = [70, 75, 80, 65]

total = sum(marks)

count = len(marks)

average = total / count

print("Average Marks:", average)
```
