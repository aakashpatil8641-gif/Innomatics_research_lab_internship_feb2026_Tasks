# Problem Statement 1:Smart Parking Lot Management System

```python
def manage_parking(capacity, logs):
    current_vehicles = 0
    peak_usage = 0

    for action in logs:
        if action == "IN":
            current_vehicles += 1
        elif action == "OUT":
           
            if current_vehicles > 0:
                current_vehicles -= 1
        
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    if current_vehicles >= capacity:
        status = "Full / Over Capacity"
    else:
        status = "Available"

    print(f"Currently Parked Vehicles: {current_vehicles}")
    print(f"Parking Status: {status}")
    
    if current_vehicles > capacity:
        print("!! ALERT: Parking capacity exceeded !!")


parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(parking_capacity, vehicle_logs)
```

#  Problem Statement 2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5   

    if traffic.lower() == "low":
        traffic_delay = 0
    elif traffic.lower() == "medium":
        traffic_delay = 10
    elif traffic.lower() == "high":
        traffic_delay = 15
    else:
        traffic_delay = 0

    if weather.lower() == "clear":
        weather_delay = 0
    elif weather.lower() == "cloudy":
        weather_delay = 5
    elif weather.lower() == "rainy":
        weather_delay = 10
    else:
        weather_delay = 0

    total_time = base_time + traffic_delay + weather_delay

    return total_time

distance = 8
traffic = "High"
weather = "Rainy"

estimate_time = estimate_delivery_time(distance, traffic, weather)
print("Estimated Delivery Time:", estimate_time, "minutes")
```

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def analyze_seat_occupancy(total_seats, booked_seats):
    booked_count = sum(booked_seats)
    
    occupancy = (booked_count / total_seats) * 100
    
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    
    if occupancy >= 90:
        suggestion = "Consider opening an additional show."
    else:
        suggestion = "No additional show required."
    
    return occupancy, status, suggestion

total_seats = 200
booked_seats = [1] * 150  

result = analyze_seat_occupancy(total_seats, booked_seats)

print("Occupancy:", result[0], "%")
print("Show Status:", result[1])
print("Suggestion:", result[2])
```

#  Problem Statement 4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    
    average = total / len(cpu_readings)
    
    if average < 50:
        status = "Normal"
    elif 50 <= average <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    return average, status

cpu_readings = [45, 60, 70, 85, 90]

avg, server_status = classify_server_load(cpu_readings)

print("Average CPU Load:", avg, "%")
print("Server Status:", server_status)
```

#  Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def monitor_resource_usage(resource_usage):
    
    overused = []
    
    threshold = 8
    
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)
    
    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"
    
    return overused, alert

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

overused_resources, energy_alert = monitor_resource_usage(usage)

print("Overused Resources:", ", ".join(overused_resources))
print("Energy Alert:", energy_alert)
```

#  Problem Statement 6: Online Event Registration Capacity Controller

```python
def manage_event_registrations(capacity, total_registrations):
    
    confirmed = 0
    waitlisted = 0
    
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    
    return confirmed, waitlisted, status

event_capacity = 100
registrations = 105

confirmed_reg, waitlisted_users, registration_status = manage_event_registrations(
    event_capacity, registrations
)

print("Confirmed Registrations:", confirmed_reg)
print("Waitlisted Users:", waitlisted_users)
print("Registration Status:", registration_status)
```
