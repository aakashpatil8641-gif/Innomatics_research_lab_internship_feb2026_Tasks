```python
username = "admin"
password = "1234"

confirm_username = input("Username :")
confirm_password = input("Password :")

if (confirm_username == username):
    print("Login Successfully")
else :
    print("wrong username")

if (confirm_password == password):
    print("Done")

else :
      print("wrong password")
```

```python
marks = [45, 78, 90, 33,60]

pass_count = 0
fail_count = 0

for m in marks:
    if m >= 50:
        pass_count += 1
    else :
        fail_count += 1



print("Total students:", len(marks))
print("Pass students:", pass_count)
print("Fail students:", fail_count)
        
```

```python
names = ["Alice", "bob", "CHARLIE"]

Clean_names = []

for name in names:
    clean_name = name.strip().lower()
    Clean_names.append(clean_name)

print("Clean Names:",Clean_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for text in messages:
    size = len(text)
    print("message:", text)
    print("length:", size)

    if (size > 10):
        print("long message")
    else:
        print("normal message")

    print("------------------")
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_total = 0

for item in logs:
    if item == "ERROR":
        error_total = error_total +1

print("Number of Error messages:",error_total)
```
