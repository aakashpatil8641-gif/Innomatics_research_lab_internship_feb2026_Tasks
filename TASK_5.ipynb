# Problem 1: Social Media â€“ Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0
    
    for like in likes_list:
        total_likes += like
    
    print("Total Likes:", total_likes)
    
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


likes = [200, 300, 150, 400]   # Total = 1050

analyze_engagement(likes)
```

# Problem 3: Agriculture â€“ Rainfall Adequacy Checker

```python
def analyze_rainfall(rainfall_data, required_level):
    
    total = 0
    
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    print("Average Rainfall:", int(average))
    
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


rainfall = [60, 70, 80, 78]   
required = 70

analyze_rainfall(rainfall, required)
```

# Problem 4: Social Media â€“ Duplicate Account Detection

```python
def detect_duplicates(usernames):
    
    unique_users = set(usernames)   
    
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = ["aakash", "rahul", "sneha", "aakash"]

detect_duplicates(users)
```

# Problem 6: Agriculture â€“ Premium Crop Price Filter

```python
def filter_premium_crops(prices):
    
    premium_crops = []
    
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)

crop_prices = [1500, 2500, 1800, 3200, 1900]

filter_premium_crops(crop_prices)
```

# Problem 9: E-Learning â€“ Student Attendance Eligibility System

```python
def check_exam_eligibility(attendance_list):
    
    total_classes = len(attendance_list)
    present_count = 0
    
    for status in attendance_list:
        if status == "P":
            present_count += 1
    
    attendance_percentage = (present_count / total_classes) * 100
    
    print("Attendance Percentage:", attendance_percentage)
    
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance = ["P", "P", "A", "P", "P"]  
check_exam_eligibility(attendance)
```

# Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    
    bill = 0
    
    if units <= 100:
        bill = units * 3
    
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    
    return bill, usage_status

units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: â‚¹", total_bill)
print("Usage Level:", status)
```
